```python
import matplotlib.pyplot as plt
import numpy as np

# Data jumlah gambar awal dan setelah ditambah 8
labels = [chr(i) for i in range(65, 91)]  # Huruf A-Z
jumlah_awal = [53, 34, 10, 10, 11, 8, 12, 6, 6, 8, 12, 16, 10, 10, 7, 9, 13, 9, 10, 8, 10, 10, 9, 8, 13, 10]
tambahan = 5
jumlah_baru = [x + tambahan for x in jumlah_awal]

# Plot
x = np.arange(len(labels))  # Lokasi kelompok
width = 0.35  # Lebar bar

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(x - width/2, jumlah_awal, width, label='Minggu Pertama')
bars2 = ax.bar(x + width/2, jumlah_baru, width, label='Minggu Kedua')

# Labeling
ax.set_xlabel('Huruf')
ax.set_ylabel('Jumlah Gambar')
ax.set_title('Perbandingan Jumlah Gambar Minggu Pertama dan Minggu Kedua')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Menambahkan nilai di atas bar
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{int(height)}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Offset text
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data jumlah gambar awal dan setelah ditambah sesuai aturan
labels = [chr(i) for i in range(65, 91)]  # Huruf A-Z
jumlah_awal = [53, 34, 10, 10, 11, 8, 12, 6, 6, 8, 12, 16, 10, 10, 7, 9, 13, 9, 10, 8, 10, 10, 9, 8, 13, 10]

# Aturan tambahan untuk minggu kedua
tambahan_minggu_kedua = [7 if label in ['A', 'B'] else 8 for label in labels]
jumlah_minggu_kedua = [x + t for x, t in zip(jumlah_awal, tambahan_minggu_kedua)]

# Aturan tambahan untuk minggu ketiga
tambahan_minggu_ketiga = [4] * len(labels)
jumlah_minggu_ketiga = [x + t for x, t in zip(jumlah_minggu_kedua, tambahan_minggu_ketiga)]

# Plot
x = np.arange(len(labels))  # Lokasi kelompok
width = 0.25  # Lebar bar

fig, ax = plt.subplots(figsize=(14, 7))
bars1 = ax.bar(x - width, jumlah_awal, width, label='Minggu Pertama')
bars2 = ax.bar(x, jumlah_minggu_kedua, width, label='Minggu Kedua')
bars3 = ax.bar(x + width, jumlah_minggu_ketiga, width, label='Minggu Ketiga')

# Labeling
ax.set_xlabel('Huruf')
ax.set_ylabel('Jumlah Gambar')
ax.set_title('Perbandingan Jumlah Gambar Minggu Pertama, Kedua, dan Ketiga')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Menambahkan nilai di atas bar
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{int(height)}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Offset text
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Data dalam bentuk DataFrame untuk analisis
data = {
    "Huruf": labels,
    "Minggu Pertama": jumlah_awal,
    "Minggu Kedua": jumlah_minggu_kedua,
    "Minggu Ketiga": jumlah_minggu_ketiga,
    "Minggu Keempat": jumlah_minggu_keempat,
}

df = pd.DataFrame(data)

# Menambahkan kolom selisih antar minggu
df["Selisih Minggu 1-2"] = df["Minggu Kedua"] - df["Minggu Pertama"]
df["Selisih Minggu 2-3"] = df["Minggu Ketiga"] - df["Minggu Kedua"]
df["Selisih Minggu 3-4"] = df["Minggu Keempat"] - df["Minggu Ketiga"]

# Menampilkan tabel EDA
print("Exploratory Data Analysis:")
print(df)

# Plot selisih antar minggu
fig, ax = plt.subplots(figsize=(16, 8))
x = np.arange(len(labels))
width = 0.25

bars1 = ax.bar(x - width, df["Selisih Minggu 1-2"], width, label="Selisih Minggu 1-2", color='cyan')
bars2 = ax.bar(x, df["Selisih Minggu 2-3"], width, label="Selisih Minggu 2-3", color='magenta')
bars3 = ax.bar(x + width, df["Selisih Minggu 3-4"], width, label="Selisih Minggu 3-4", color='orange')

# Labeling
ax.set_xlabel('Huruf')
ax.set_ylabel('Selisih Jumlah Gambar')
ax.set_title('Selisih Jumlah Gambar Antar Minggu')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Menambahkan nilai di atas bar
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{int(height)}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Offset text
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
