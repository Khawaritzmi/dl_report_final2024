```python
import matplotlib.pyplot as plt
import numpy as np

# Data huruf A-Z
labels = [chr(i) for i in range(65, 91)]  # Huruf A-Z

# Data jumlah gambar per minggu
jumlah_awal = [53, 34, 10, 10, 11, 8, 12, 6, 6, 8, 12, 16, 10, 10, 7, 9, 13, 9, 10, 8, 10, 10, 9, 8, 13, 10]
tambahan_minggu_kedua = [7 if label in ['A', 'B'] else 8 for label in labels]
jumlah_minggu_kedua = [x + t for x, t in zip(jumlah_awal, tambahan_minggu_kedua)]
tambahan_minggu_ketiga = [4] * len(labels)
jumlah_minggu_ketiga = [x + t for x, t in zip(jumlah_minggu_kedua, tambahan_minggu_ketiga)]

# Data minggu keempat
jumlah_minggu_keempat = [73, 53, 31, 31, 30, 29, 28, 27, 26, 35, 29, 31, 28, 28, 28, 29, 33, 31, 29, 26, 28, 29, 29, 27, 33, 29]

# Plot data
x = np.arange(len(labels))  # Lokasi kelompok
width = 0.2  # Lebar bar

fig, ax = plt.subplots(figsize=(16, 8))

# Plot data per minggu dengan warna merah, hijau, kuning, dan biru
bars1 = ax.bar(x - width * 1.5, jumlah_awal, width, label='Minggu Pertama', color='red')
bars2 = ax.bar(x - width / 2, jumlah_minggu_kedua, width, label='Minggu Kedua', color='green')
bars3 = ax.bar(x + width / 2, jumlah_minggu_ketiga, width, label='Minggu Ketiga', color='yellow')
bars4 = ax.bar(x + width * 1.5, jumlah_minggu_keempat, width, label='Minggu Keempat', color='blue')

# Labeling
ax.set_xlabel('Huruf')
ax.set_ylabel('Jumlah Gambar')
ax.set_title('Perbandingan Jumlah Gambar per Minggu')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Menambahkan nilai di atas bar
for bars in [bars1, bars2, bars3, bars4]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{int(height)}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Offset text
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Dataset path
dataset_path = '/content/drive/MyDrive/Dataset Baru/Gambar'

# Menampilkan kelas dalam dataset dan mengurutkan dari A sampai Z
classes = sorted(os.listdir(dataset_path))  # Mengurutkan kelas secara alfabetis
print("Classes: ", classes)

# Inisialisasi variabel untuk menghitung total gambar
total_images = 0

# Menampilkan jumlah gambar dalam setiap kelas
for cls in classes:
    class_path = os.path.join(dataset_path, cls)
    if os.path.isdir(class_path):
        num_images = len(os.listdir(class_path))  # Menghitung jumlah gambar dalam direktori
        total_images += num_images  # Menambahkan jumlah gambar ke total
        print(f"{cls}: {num_images} gambar")
    else:
        print(f"{cls} is not a directory, skipping.")

# Menampilkan total jumlah gambar
print(f"Total gambar di seluruh dataset: {total_images} gambar")

# Mengatur ukuran figure untuk menampilkan gambar yang lebih jelas
plt.figure(figsize=(8, 16))

# Menentukan jumlah kolom untuk grid, misalnya 3 gambar per baris
num_cols = 3
num_rows = (len(classes) + num_cols - 1) // num_cols  # Menghitung jumlah baris berdasarkan jumlah kelas

# Menampilkan beberapa gambar contoh dari setiap kelas
for i, cls in enumerate(classes):
    class_path = os.path.join(dataset_path, cls)
    if os.path.isdir(class_path):
        sample_imgs = os.listdir(class_path)

        for j in range(min(1, len(sample_imgs))):  # Menampilkan maksimal 1 gambar per kelas
            sample_img = sample_imgs[j]
            img = mpimg.imread(os.path.join(class_path, sample_img))
            plt.subplot(num_rows, num_cols, i + 1)
            plt.imshow(img)
            plt.title(cls)
            plt.axis('off')
            plt.gca().set_aspect('auto')  # Menjaga rasio gambar agar tidak terdistorsi
    else:
        print(f"{cls} is not a directory, skipping.")

plt.tight_layout()
plt.show()
```

```python
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns

# Base directory of the dataset
base_dir = '/content/drive/MyDrive/Semaphore Dataset.v2i.yolov8'
train_label_dir = os.path.join(base_dir, 'train/labels')
valid_label_dir = os.path.join(base_dir, 'valid/labels')
test_label_dir = os.path.join(base_dir, 'test/labels')

# Initialize class distribution dictionary for A-Z
class_distributions = {
    'train': {chr(i): 0 for i in range(65, 91)},  # A-Z
    'valid': {chr(i): 0 for i in range(65, 91)},  # A-Z
    'test': {chr(i): 0 for i in range(65, 91)},   # A-Z
}

# Function to count class instances from label files
def count_classes(label_dir, distribution_dict):
    if not os.path.exists(label_dir):
        return
    label_files = glob.glob(os.path.join(label_dir, '*.txt'))
    for label_file in label_files:
        with open(label_file, 'r') as file:
            for line in file:
                class_id = int(line.strip().split()[0])  # Get the class ID
                class_label = chr(65 + class_id)  # Convert to A-Z letter
                if class_label in distribution_dict:
                    distribution_dict[class_label] += 1

# Count classes for each dataset (train, valid, test)
count_classes(train_label_dir, class_distributions['train'])
count_classes(valid_label_dir, class_distributions['valid'])
count_classes(test_label_dir, class_distributions['test'])

# Calculate total counts across all datasets for A-Z
totals = {chr(i): sum(d[chr(i)] for d in class_distributions.values()) for i in range(65, 91)}

# Print total counts for A-Z
print("Total counts across all datasets (A-Z):")
for category, total in totals.items():
    print(f"{category}: {total}")

# Prepare combined data for plotting
combined_counts = []
categories = []
datasets = []

# Organize data for plotting
for dataset, counts in class_distributions.items():
    for category, count in counts.items():
        datasets.append(dataset.capitalize())  # Train, Valid, Test
        categories.append(category)  # A, B, C, ..., Z
        combined_counts.append(count)

# Plot class distribution across datasets
plt.figure(figsize=(16, 8))
sns.barplot(x=datasets, y=combined_counts, hue=categories)
plt.xlabel("Dataset")
plt.ylabel("Count")
plt.title("Class Distribution Across Train, Validation, and Test (A-Z)")
plt.legend(title="Class")
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns

# Base directory of the dataset
base_dir = '/content/drive/MyDrive/Semaphore Dataset.v2i.yolov8'
train_label_dir = os.path.join(base_dir, 'train/labels')
valid_label_dir = os.path.join(base_dir, 'valid/labels')
test_label_dir = os.path.join(base_dir, 'test/labels')

# Initialize class distribution dictionary for A-Z
class_distributions = {
    'train': {chr(i): 0 for i in range(65, 91)},  # A-Z
    'valid': {chr(i): 0 for i in range(65, 91)},  # A-Z
    'test': {chr(i): 0 for i in range(65, 91)},   # A-Z
}

# Function to count class instances from label files
def count_classes(label_dir, distribution_dict):
    if not os.path.exists(label_dir):
        return
    label_files = glob.glob(os.path.join(label_dir, '*.txt'))
    for label_file in label_files:
        with open(label_file, 'r') as file:
            for line in file:
                class_id = int(line.strip().split()[0])  # Get the class ID
                class_label = chr(65 + class_id)  # Convert to A-Z letter
                if class_label in distribution_dict:
                    distribution_dict[class_label] += 1

# Count classes for each dataset (train, valid, test)
count_classes(train_label_dir, class_distributions['train'])
count_classes(valid_label_dir, class_distributions['valid'])
count_classes(test_label_dir, class_distributions['test'])

# Calculate total counts across all datasets for A-Z
totals = {chr(i): sum(d[chr(i)] for d in class_distributions.values()) for i in range(65, 91)}

# Print total counts for each dataset (train, valid, test) and overall (A-Z)
print("Class Distribution in Each Dataset:")
for dataset in ['train', 'valid', 'test']:
    print(f"\n{dataset.capitalize()} Dataset:")
    for category, count in class_distributions[dataset].items():
        print(f"{category}: {count}")

print("\nTotal counts across all datasets (A-Z):")
for category, total in totals.items():
    print(f"{category}: {total}")

# Prepare combined data for plotting
combined_counts = []
categories = []
datasets = []

# Organize data for plotting
for dataset, counts in class_distributions.items():
    for category, count in counts.items():
        datasets.append(dataset.capitalize())  # Train, Valid, Test
        categories.append(category)  # A, B, C, ..., Z
        combined_counts.append(count)

# Plot class distribution across datasets
plt.figure(figsize=(16, 8))
ax = sns.barplot(x=datasets, y=combined_counts, hue=categories)
plt.xlabel("Dataset")
plt.ylabel("Count")
plt.title("Class Distribution Across Train, Validation, and Test (A-Z)")
plt.legend(title="Class")
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Add numbers on top of each bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                fontsize=10, color='black',
                xytext=(0, 5), textcoords='offset points')

plt.show()
```
