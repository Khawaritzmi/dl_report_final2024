```python
import requests
import os

IMAGE_URL = "https://www.thiswaifudoesnotexist.net/example-{num}.jpg"

num_images = 42000
output_folder = "/content/drive/My Drive/waifu_images"

os.makedirs(output_folder, exist_ok=True)

image_num = 8939
for i in range(40000, num_images + 1):
    try:
        url = IMAGE_URL.format(num=i)
        response = requests.get(url, stream=True)

        if response.status_code == 200:
            image_path = os.path.join(output_folder, f"ai_twdne_{image_num}.jpg")
            with open(image_path, "wb") as file:
                file.write(response.content)
            print(f"Downloaded {image_path}")
            image_num += 1
        else:
            print(f"Failed to download image {i}")
    except Exception as e:
        print(f"An error occurred: {e}")

print("Download complete.")
```
